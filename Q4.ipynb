```python
from myqrsteps_skeleton import myqrsteps
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Load the file and convert it to a matrix
matrix = np.loadtxt('MYMATRIX.txt')
```

```python
U, S, Vt = np.linalg.svd(matrix, full_matrices=False)
```

```python


plt.plot(S, 'o-', label='Singular Values')
plt.title('Singular Values (Regular Scale)')
plt.xlabel('Index')
plt.ylabel('Singular Value')
plt.legend()
plt.grid(True)
plt.show()


plt.plot(S, 'o-', label='Singular Values (Log Scale)')
plt.yscale('log')
plt.title('Singular Values (Log Scale)')
plt.xlabel('Index')
plt.ylabel('Singular Value (Log Scale)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

approximations = [1, 3, 5, 7]
def low_rank_approximation(U, S, Vt, rank):
    S_approx = np.zeros((len(S), len(S)))
    np.fill_diagonal(S_approx, S[:rank])
    return np.dot(U[:, :rank], np.dot(S_approx[:rank, :rank], Vt[:rank, :]))
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

for i, rank in enumerate(approximations):
    low_rank_matrix = low_rank_approximation(U, S, Vt, rank)
    axes[i].imshow(low_rank_matrix, cmap='gray')
    axes[i].set_title(f'Rank-{rank} Approximation')
    axes[i].axis('off')
plt.show()
```
